```python
# Importação das bibliotecas necessárias para manipulação dos dados
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Leitura dos arquivo com os dataset
g1 = pd.read_csv('datasets/kaggle_diogo_caliman_historico_materias.csv')
folha_sp = pd.read_csv('datasets/kaggle_marlesson_noticias_folha_sp.csv')
```

```python

def bargraph(data, x, y, ylabel, xlabel):
    ax = sns.barplot(data = data, x = x, y = y)
    ax.set(ylabel=ylabel)
    ax.set(xlabel=xlabel)
```

```python
plt.figure(figsize=(15,10))
bargraph(data = g1, x = g1.assunto.unique(), y = g1.assunto.value_counts(), xlabel='Categorias', ylabel='Contagem')
```

```python
plt.figure(figsize=(15,10))
bargraph(data = folha_sp, x = folha_sp.category.unique(), y = folha_sp.category.value_counts(), xlabel='Categorias', ylabel='Contagem')
```

```python
folha_sp.category.value_counts().to_csv('teste.csv', sep=';')
```
